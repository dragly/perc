# Flux through a percolating cluster #

About a year ago, I wrote about [how you can construct a percolation system in Python](http://dragly.org/2013/03/25/working-with-percolation-clusters-in-python/).

In this notebook, I will show how to calculate the flux through a percolating cluster.
The flux arises because we apply a pressure $\Delta p$ across the system.
We will also find the resulting pressure in each site.
As a bonus we'll get pretty-looking figures like the one below.

<img src="http://dragly.org/wp-content/uploads/2014/12/flux-e1418393953867.png" />

This shows the magnitude of the flux through each site.

First, we'll import the necessary packages:

```python
%pylab inline
from scipy.ndimage import measurements
from scipy.sparse import spdiags, dia_matrix, coo_matrix
from scipy.sparse.linalg import spsolve
```

Let's start out by setting up the percolation system similarly to what we did in the link above.
To obtain a percolating cluster, we need the probability to be just high enough for this to happen.
The value $p = p_c = 0.59275$ will do. 
Because we work with a finite system, we also need to add a while-loop to ensure that we have obtained a percolating cluster before continuing:

```python
found_percolating_cluster = False
while not found_percolating_cluster:
    print "Trying to create a percolating cluster..."
    nx = ny = 128
    random_matrix = rand(ny,nx)
    probability_of_site_being_occupied = 0.59275
    
    # Find the occupied sites and convert the result to floating point numbers
    occupation_matrix = (random_matrix < probability_of_site_being_occupied).astype(float)
    
    # Find and label all the clusters
    cluster_label_matrix, largest_cluster_number = measurements.label(occupation_matrix)
    
    # Iterate over all the labeled clusters (0 is the unoccupied sites)
    for i in range(1, largest_cluster_number + 1):
        # Calculate the bounding box of the cluster i
        sliced = measurements.find_objects(cluster_label_matrix == i)
        if(len(sliced) > 0):
            sliceY = sliced[0][0]
            sliceX = sliced[0][1]
            # If the cluster spans the whole system in the x-direction, it is percolating
            if sliceX.start == 0 and sliceX.stop == nx:
                found_percolating_cluster = True
                percolating_cluster = (cluster_label_matrix == i).astype(float)
                print "Success!"

figure(figsize=(8,8))
imshow(percolating_cluster, cmap="GnBu_r", interpolation='none')
colorbar()
# color unoccupied areas as black
masked_data = zeros_like(occupation_matrix)
masked_data = np.ma.masked_where(occupation_matrix > 0, masked_data)
imshow(masked_data, interpolation="none", cmap="hot")
title("Percolating cluster")
```

We are now going to introduce flow through the percolating cluster.
We assume that the flow is determined by [Fick's second law of diffusion](http://en.wikipedia.org/wiki/Fick%27s_laws_of_diffusion):

$$\frac{\partial p}{\partial t} = -D \nabla^2 p$$

We could solve this equation in many ways, for instance using a finite difference scheme.
However, we will solve this for the steady state.
That is, when the time evolution of the pressure is zero,

$$\frac{\partial p}{\partial t} = 0.$$

Resulting in

$$0 = -D \nabla^2 p.$$

Note that this also is the same as saying that the net flux into a site should equal the net flux out of that same site.

We discretize the Laplace operator using a second order finite difference scheme.

$$
0
=
-
\frac{D}
{\Delta y^2}
\left( 
p_{i-1,j} - 2p_{ij} + p_{i+1,j}
\right)
-
\frac{D}
{\Delta x^2}
\left( 
p_{i,j-1} - 2p_{ij} + p_{i,j+1}
\right)
$$

Before we can solve this, we need to decide on some boundary conditions.
We choose an open boundary with Dirichlet boundary conditions on the left and right sides of the system.
That means that we explicitly fix the values on the boundary:

\begin{align}
p_{i,0} &= b_i \\
p_{i,n_x - 1} &= c_i
\end{align}

We choose a closed boundary with von Neumann boundary conditions on the top and bottoms sides of the system.
This means that the derivative of the pressure is zero over the boundary:

\begin{align}
\frac{\partial}{\partial y}
p_{0,j} &= 0 
\\
\frac{\partial}{\partial y}
p_{n_y - 1,j} &= 0
\end{align}

We will implement this by a simple forward and backward difference:

\begin{align}
p_{0,j} - p_{-1,j} &= 0 
\\
p_{n_y, j} - p_{n_y - 1,j} &= 0
\end{align}

Further, note that we have some sites that are unavailable.
We need to decide what to with these.

First of all, the pressure in unoccupied sites should be fixed to zero.
Further, the flow in and out of such a site is zero.
We should therefore have a zero gradient at the boundary of each such site $(i,j)$.
The boundary is defined at grid points $\left (i \pm \frac{1}{2}, j \right)$ and $\left (i, j\pm \frac{1}{2} \right)$.
We should therefore have

\begin{align}
\frac{\partial}{\partial y} p_{i + \frac{1}{2}, j} &= 0 \\
\frac{\partial}{\partial y} p_{i - \frac{1}{2}, j} &= 0 \\
\frac{\partial}{\partial x} p_{i, j + \frac{1}{2}} &= 0 \\
\frac{\partial}{\partial x} p_{i, j - \frac{1}{2}} &= 0
\end{align}

where $(i,j)$ is an occupied site.

In discretized terms, we get

\begin{align}
p_{i + 1, j} - p_{i, j} &= 0 \\
p_{i, j} - p_{i - 1, j} &= 0 \\
p_{i, j + 1} - p_{i, j} &= 0 \\
p_{i, j} - p_{i, j - 1} &= 0 \\
\end{align}

To achieve this, we introduce a new variable $z_{ij}$. This is defined such that $z_{ij} = 0$ if $(i,j)$ is unavailable and $z_{ij} = 1$ otherwise.
This allows us to rewrite the above continuity equation:

$$
0
=
-
\frac{D}
{2\Delta y^2}
\left( 
z_{i-1,j} z_{ij} (p_{i-1,j} - p_{ij}) + z_{i+1,j} z_{ij}(p_{i+1,j} - p_{ij})
\right)
-
\frac{D}
{2\Delta x^2}
\left( 
z_{i,j-1} z_{ij} (p_{i,j-1} - p_{ij}) + z_{i,j+1} z_{ij} (p_{i,j+1} - p_{ij})
\right)
$$

To make this into a matrix problem, we factor out the pressures $p_{ij}$:

\begin{align}
0
=
&
-
\frac{D}{2\Delta y^2} z_{i-1,j} z_{ij} p_{i-1,j} 
\\
&
- 
\frac{D}{2\Delta y^2} z_{i+1,j} z_{ij} p_{i+1,j}
\\
&
+
\left (
\frac{D}{2\Delta y^2} (z_{i-1,j} + z_{i+1,j})
+
\frac{D}{2\Delta x^2} (z_{i,j-1} + z_{i,j+1})
\right)
z_{ij}
p_{ij}
\\
&
-
\frac{D}{2\Delta x^2} z_{i,j-1} z_{ij} p_{i,j-1} 
\\
&
-
\frac{D}{2\Delta x^2} z_{i,j+1} z_{ij} p_{i,j+1}
\end{align}

This can be written as a matrix problem

$$\mathbf A \mathbf p = \mathbf b.$$

Where the main diagonal elements are

$$
A_{in_x + j, in_x + j} 
= 
z_{ij}
\left (
\frac{D}{2\Delta y^2} (z_{i-1,j} + z_{i+1,j})
+
\frac{D}{2\Delta x^2} (z_{i,j-1} + z_{i,j+1})
\right)
$$

and equivalently for the two upper and the two lower diagonals.

The boundary conditions impact elements such as $z_{-1,0}$ and are handled using what is known as ghost sites.
These are just temporary objects to make the equations easier to works with.
In principle, this is done by expanding the whole system with one extra row or column of sites in each direction so that we have $(n_y + 2) \times (n_x + 2)$ sites to work with.

These sites are used to set up the necessary equations for the boundary conditions above.

Note that we don't have to use ghost sites.
We could rather incorporate the boundary conditions directly into our existing equations.
However, this tends to be a bit more work, so many prefer to use ghost sites instead.

We start out by defining the diffusion constant, our grid size, the number of cells including grid cells, etc.

```python
D = 1.0
dx = 1.0
dy = 1.0

neq_inner = ny * nx

# number of cells, including ghost cells
nyg = ny + 2
nxg = nx + 2
neq_total = (nyg)*(nxg)

# indexes for the inner cells
idy = slice(1, -1)
idx = slice(1, -1)

# the whole system without ghost cells
z = percolating_cluster

# the whole system with ghost cells
zg = zeros((nyg, nxg))

# we set the ghost cells to have the same values
# as the system boundary
zg[0, idx] = 0 # von Neumann
zg[-1, idx] = 0 # von Neumann
zg[idy, 0] = 1 # Dirichlet
zg[idy, -1] = 1 # Dirichlet 

# the inner points are determined by the z values
zg[idy, idx] = percolating_cluster
```

Next, we set up the equation system.
This is done by thinking of the factors in front of the pressure $p_{ij}$ as the diagonals in the $(n_x + 2) (n_y + 2) \times (n_x + 2) (n_y + 2) $ matrix $\mathbf A$.
However, for simplicity, we can work with each diagonal as a $n_x \times n_y$ matrix before reshaping it into a 1D vector and putting it on the right diagonal of $\mathbf A$.
This makes it a bit easier to keep track of the individual contributions to $\mathbf A$.

```python
main_diagonal = zeros_like(zg)

# for all except the ghost cells
main_diagonal[1:-1, 1:-1] += zg[1:-1, 1:-1] * D / dy * (zg[:-2, 1:-1] + zg[2:, 1:-1])
main_diagonal[1:-1, 1:-1] += zg[1:-1, 1:-1] * D / dx * (zg[1:-1, :-2] + zg[1:-1, 2:])

# the ghost cells are used for the boundary condition equations
main_diagonal[0, :] = 1
main_diagonal[-1, :] = 1
main_diagonal[:, 0] = 1
main_diagonal[:, -1] = 1

# all sites that are not occupied should be set to zero, 
# so the main diagonal should be one to set up these equations
main_diagonal[where(main_diagonal == 0)] = 1.0

up_diagonal = zeros_like(zg)
up_diagonal[idy, idx] = -D / dy * zg[idy, idx] * zg[:-2, idx]
# von Neumann
up_diagonal[-1, idx] = -zg[-1, idx]

down_diagonal = zeros_like(zg)
down_diagonal[idy, idx] = -D / dy * zg[idy, idx] * zg[2:, idx]
# von Neumann
down_diagonal[0, idx] = -zg[0, idx]

left_diagonal = zeros_like(zg)
left_diagonal[idy, idx] = -D / dx * zg[idy, idx] * zg[idy, :-2]
# Dirichlet
left_diagonal[idy, -1] = 0

right_diagonal = zeros_like(zg)
right_diagonal[idy, idx] = -D / dx * zg[idy, idx] * zg[idy, 2:]
# Dirichlet
right_diagonal[idy, 0] = 0

# Convert from matrices to vectors
main_diagonal_vector = main_diagonal.reshape(neq_total)
up_diagonal_vector = up_diagonal.reshape(neq_total)
down_diagonal_vector = down_diagonal.reshape(neq_total)
left_diagonal_vector = left_diagonal.reshape(neq_total)
right_diagonal_vector = right_diagonal.reshape(neq_total)
```

Now that we have all the diagonals ready as 1D vectors, we need to put them on the right place in $\mathbf A$.
This is done using the spdiags function.
To create a 10x10 diagonal matrix with the numbers 1 to 10 on the main diagonal, we could just write spdiags(arange(1, 11), 0, 10, 10).
Note that the behavior of spdiags is a bit unintuitive for the other diagonals. Feel free to play around with spdiags to get a better understanding of it.

```python
A = spdiags(main_diagonal_vector, 0, neq_total, neq_total)
A = A + spdiags(up_diagonal_vector[nxg:], -nxg, neq_total, neq_total)
A = A + spdiags(down_diagonal_vector, -nxg, neq_total, neq_total).T
A = A + spdiags(left_diagonal_vector[1:], -1, neq_total, neq_total)
A = A + spdiags(right_diagonal_vector, -1, neq_total, neq_total).T
```

We need to set up the right hand side of the matrix equation system, $\mathbf b$.
To introduce flow from the left, we set this to 1 on each ghost site that is available on the left side.
Afterwards, we reshape the matrix to make $\mathbf b$ a 1D vector.

```python
# input from the left side where there are openings
b_matrix = zeros((nyg, nxg))
b_matrix[idy, 0] = 1.0 * zg[idy, 0]
b = b_matrix.reshape(neq_total)
```

All that remains now is to solve the equation system and plot the resulting pressure in each site.

```python
phi = spsolve(A, b)

# Convert back from vector to matrix
phi = phi.reshape((nyg, nxg))

figure(figsize=(8,8))
phi = phi[idx, idy]
imshow(phi, interpolation="none", cmap="GnBu_r")
print phi.max()
print phi.min()
colorbar()
```

Let's calculate the flow.
This is found by taking the pressure difference between a site and its neighbors.
The net flow into a site is of course zero, otherwise we would have accumulation of pressure.
This is in other words not an interesting measure.
What we need is therefore the amount of flow through a site, which is found by summing the absolute value of the flow over the edges of a site: 

```python
flow = zeros((ny, nx))
flow[1:, :] += abs(phi[:-1, :] - phi[1:, :]) * z[1:, :] * z[:-1, :]
flow[:-1, :] += abs(phi[1:, :] - phi[:-1, :]) * z[:-1, :] * z[1:, :]
flow[:, 1:] += abs(phi[:, :-1] - phi[:, 1:]) * z[:, 1:] * z[:, :-1]
flow[:, :-1] += abs(phi[:, 1:] - phi[:, :-1]) * z[:, :-1] * z[:, 1:]

figure(figsize=(8,8))
imshow(flow, interpolation="none", cmap="GnBu_r")
colorbar()
```

That's it! I will leave it up to you to replace constants such as $D$ and $\Delta x$ with real values and add proper units to the results.
